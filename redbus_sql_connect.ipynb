```python
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_buses_1=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Kerala Bus details.csv")
df_buses_2=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Andhra bus details")
df_buses_3=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Telangana bus details")
df_buses_4=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Rajasthan bus details")
df_buses_5=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\HP bus details")
df_buses_6=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Assam bus details")
df_buses_7=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\UP bus details")
df_buses_8=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\WB bus details")
df_buses_9=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Bihar bus details")
df_buses_10=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\streamlit\Punjab bus details")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

```python
Final_df.info()
```

```python
Final_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
Final_df
```

```python
Final_df.info()
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
```

```python
Final_df["Price"].fillna(0,inplace=True) 
```

```python
Final_df["Price"]
```

```python
#convert seat availability to numeric
Final_df["Seats_Available"]=Final_df["Seats_Available"].str.replace("Seats available","")
Final_df["Seats_Available"]=Final_df["Seats_Available"].str.strip()
Final_df["Seats_Available"]=Final_df["Seats_Available"].str.split().str[0]
Final_df["Seats_Available"].fillna(0,inplace=True) 
```

```python
Final_df["Seats_Available"]
```

```python
Final_df["Seats_Available"] = pd.to_numeric(Final_df["Seats_Available"], errors='coerce')
Final_df["Seats_Available"].fillna(0,inplace=True) 
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True) 
```

```python
Final_df["Ratings"]
```

```python
Final_df.info()
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
Final_df
```

```python
Final_df.to_csv("Redbus_Final_details.csv")
```
